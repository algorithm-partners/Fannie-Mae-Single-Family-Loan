```python
# df = pd.read_csv('Acquisition_2000Q1.txt', sep='|',header = None)
# df2 = pd.read_csv('Acquisition_2001Q2.txt', sep = '|', header = None)

catalog = list()
for i in range(18):
    for j in range(1,5):
        catalog.append('Acquisition_' + str(2000+i) + 'Q' + str(j) + '.txt')
catalog.append('Acquisition_2018Q1.txt')

# df = pd.DataFrame()
# for item in catalog:
#     df2 = pd.read_csv(item, sep = '|', header = None, error_bad_lines = False)
#     df = pd.concat([df,df2])
```

```python
dfl = list()
for i in range(len(catalog)):
    df = pd.read_csv(catalog[i], sep='|',header = None, error_bad_lines = False)
    dfl.append(df)

    
```

```python
dft = pd.concat(dfl, ignore_index= True)
```

```python
dft.columns = ["LOAN_ID", "ORIG_CHN", "Seller.Name", "ORIG_RT", "ORIG_AMT", "ORIG_TRM", "ORIG_DTE"
                           ,"FRST_DTE", "OLTV", "OCLTV", "NUM_BO", "DTI", "CSCORE_B", "FTHB_FLG", "PURPOSE", "PROP_TYP"
                           ,"NUM_UNIT", "OCC_STAT", "STATE", "ZIP_3", "MI_PCT", "Product.Type", "CSCORE_C", "MI_TYPE", "RELOCATION_FLG"]
```

```python
# df = dft[['LOAN_ID','ORIG_CHN','ORIG_RT','ORIG_AMT','ORIG_DTE','OLTV','DTI','CSCORE_B','PURPOSE']]
```

```python
date = dft['ORIG_DTE']
```

```python
year = date.str.replace('\d\d/','')
```

```python
month = date.str.replace('/\d\d\d\d','')
```

```python
dft['year'] = year
```

```python
dft['month'] = month
```
